“When Logic Changes: Exploring Non-Monotonic Reasoning and Argumentation”


Part I. Conceptual Understanding
1. Non-monotonic reasoning is a type of logic where adding new information can cause you to retract or change previous conclusions.

2. Non-monotonic reasoning allows conclusions to change or be retracted when new information is added, while monotonic reasoning ensures that once a conclusion is reached, it remains true forever, regardless of any new information.

3. We first assume our friend will be at the party, but we must change that conclusion when we receive a new text saying they're sick and can't come.

4. A default rule is a general assumption used to make a conclusion when you don't have complete information, which you are prepared to change if an exception comes up.

5. Argumentation frameworks help AI systems decide between conflicting rules by modeling them as a network of arguments and attacks, allowing the system to determine which arguments are defended and logically acceptable based on the attack structure.






Part II. Laboratory Application
Task 1: Belief Revision Simulation

```
# Belief Revision Simulation

# Step 1: Initial rule
print("Rule: If an animal is a bird, assume it can fly.\n")

# Step 2: Get user input
animal = input("Input: ").strip().lower()

# Step 3: Start reasoning
print("\nReasoning:")

# Define bird exceptions
exceptions = ["penguin", "ostrich", "kiwi"]

# Step 4: Apply reasoning process
print(f"{animal.capitalize()}s are birds.")

if animal in exceptions:
    print(f"However, {animal}s do not fly.")
    conclusion = f"{animal}s cannot fly."
else:
    print(f"So, {animal}s can fly.")
    conclusion = f"{animal}s can fly."

# Step 5: Display conclusion
print("\nConclusion:", conclusion)
```
